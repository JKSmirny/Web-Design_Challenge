```python
Data=pd.read_csv("Resources/cities.csv")
pd.set_option('colheader_justify', 'center')
Data.head()
```

```python
#Use df.columns command to drop columns
weather_data = Data.drop(Data.columns[[0, 6]], axis=1)
weather_data.head()
```

```python
#rename columns 
data_weather = weather_data.rename(columns={'Temperature': 'Temp_F', 'Humidity': 'Humidity, %','wind_speed': 'Wind (mph)', 'date_norm': 'Date', 'Cloudiness': 'Cloudiness, %'})
data_weather.head()
```

```python
#converting <float> to <int> to remove decimals
#data_weather.Longitude = data_weather.Longitude.apply(int)
#data_weather.Latitude = data_weather.Latitude.apply(int)
data_weather.Longitude = round(data_weather.Longitude,2)
data_weather.Latitude = round(data_weather.Latitude,2)
data_weather.Temp_F = round(data_weather.Temp_F,2)
data_weather
```

```python
data_weather["Temp_C"] = data_weather["Temp_F"]-32
data_weather["Temp_C"] = data_weather["Temp_C"]/ 1.8
#data_weather.Temp_C = data_weather.Temp_C.apply(int)
data_weather.Temp_C = round(data_weather.Temp_C,2)
data_weather.head()
```

```python
data_weather["Wind (kmh)"] = round(data_weather["Wind (mph)"]*0.621371,2)
data_weather.head()
```

```python
cols = list(data_weather.columns.values)
cols
```

```python
data_weather = data_weather[['City', 'Country', 'Longitude', 'Latitude', 'Date', 'Temp_F', 'Temp_C', 'Wind (mph)', 'Wind (kmh)', 'Humidity, %', 'Cloudiness, %']]
data_weather.head()
```

```python
data_weather = pd.DataFrame(data_weather)
pd.set_option('colheader_justify', 'center')
data_weather.head()
```
